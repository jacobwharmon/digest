```python
import os
import pandas as pd
from datetime import date, datetime
from dateutil import relativedelta

construction_path = "raw/construction"

# Creates tuple for the construction filepath and its Mon_YYYY formatted column name
construction_files = [(f'{construction_path}/{file}', date(int(file[0:4]), int(file[4:6]), 1).strftime('%b_%Y')) for file in sorted(os.listdir(construction_path))[:-1]]

# Final file in directory is the most current, and is named totsa.xls, so we need it to give it a Mon_YYYY formatted column name as well
one_month = relativedelta.relativedelta(months=1)
construction_files.append(('raw/construction/totsa.xls',
                           (datetime.strptime(construction_files[-1][1],"%b_%Y") + one_month).strftime('%b_%Y')))
```

```python

# Read in all 215 (2004-2021)
df = pd.read_excel(construction_files[82], 1) #1-82 work for this. 
df = df[['Unnamed: 0','Unnamed: 1']].dropna().rename(columns={'Unnamed: 0':'category','Unnamed: 1':'expenditure_millions'})
df = df.set_index('category').drop(index='Type of Construction')
df
# Join these all together with date as the column name
```
